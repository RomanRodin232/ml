### 7. Создайте новый датафрейм, удалив все столбцы, связанные с семьей и происхождением студента
<br>
<p style = "font-size: 16px"> Уберем также столбцы относящиеся к другому заданию</p>

```python
df_copy = df.copy()
drop_cols = ["famsize", "famrel", "Mjob", "Fjob", "Pstatus", "Medu", "Fedu", "romantic", "internet", "higher", "nursery", "activities", "paid", "famsup", "schoolsup", "failures", "studytime", "traveltime", "guardian", "reason"]
df_copy.drop(drop_cols, axis=1, inplace=True)
print(df_copy.head())
```

### 8. По школам определите 15 студентов, имеющих наихудшую успеваемость, найдите для выделенной группы средние значения показателей health, Dalc и Walc

```python
for col in df_copy["school"].unique():
    print("Список студентов из школы:", col)
    school_data = df_copy[df_copy['school'] == col].copy()
    school_data.sort_values(by="G3", inplace=True)
    worst_stud = school_data.head(15)
    print(worst_stud)
    print(f"Среднее значение показателя health: {worst_stud['health'].mean()}")
    print(f"Среднее значение показателя Dalc: {worst_stud['Dalc'].mean()}")
    print(f"Среднее значение показателя Walc: {worst_stud['Walc'].mean()}")
    print("\n")
```

### 9. Определите 15 студентов, имеющих наилучшую успеваемость, найдите для выделенной группы средние значения показателей health, Dalc и Walc

```python
for col in df_copy["school"].unique():
    print("Список студентов из школы:", col)
    school_data = df_copy[df_copy['school'] == col].copy()
    school_data.sort_values(by="G3", inplace=True, ascending=False)
    worst_stud = school_data.head(15)
    print(worst_stud)
    print(f"Среднее значение показателя health: {worst_stud['health'].mean()}")
    print(f"Среднее значение показателя Dalc: {worst_stud['Dalc'].mean()}")
    print(f"Среднее значение показателя Walc: {worst_stud['Walc'].mean()}")
    print("\n")
```

### 10. Имеются ли в наборе лица, употребляющие спиртные напитки нелегально (возраст до 17 лет)?
<br>
<p style = "font-size: 16px"> Будем считать что при значении показателя Dalc и Walc = 1 студент не употребляет алкоголь
</p>

```python
underage_students = df_copy[(df_copy['age'] < 17) & ((df_copy['Dalc'] > 1) | (df_copy['Walc'] > 1))].copy()
print("Количество студентов нелегально употребляющие спиртные напитки:", len(underage_students))
```

### 11. Определите число таких лиц в каждой школе. Кого больше, юношей или девушек?

```python
for col in df_copy["school"].unique():
    print("Список студентов из школы:", col)
    underage_students_school = underage_students[underage_students['school'] == col].copy()
    print("Количество лиц в этой школе:", len(underage_students_school))
    m_cnt = len(underage_students_school[underage_students_school['sex'] == "M"])
    f_cnt = len(underage_students_school[underage_students_school['sex'] == "F"])
    print(f"Юношей: {m_cnt}\nДевушек: {f_cnt}")
    print("Юношей больше\n") if m_cnt > f_cnt else print("Девушек больше\n") if m_cnt < f_cnt else print("Одинаковое количество\n")
```

### 12. В какой школе наименьший процент студентов, злоупотребляющих алкоголем?
<br>
<p style = "font-size: 16px"> Будем считать что злоупотребление алкоголем считается при значении показателя Dalc или Walc > 3
</p>

```python
for col in df_copy["school"].unique():
    print("Список студентов из школы:", col)
    total_students = df_copy[df_copy['school'] == col].copy()
    alco_students = total_students[(total_students['Dalc'] > 3) | (total_students['Walc'] > 3)]
    percent_alco = len(alco_students)/len(total_students) * 100
    print(f"Процент студентов, злоупотребляющих алкоголем {percent_alco}\n")
```

<p style = "font-size: 16px"> Исходя из результатов (21.739 > 20.343) можно увидеть что в школе MS процент студентов больше.

</p>

### 13. Связаны ли каким-то образом избыток свободного времени и частые встречи с друзьями со злоупотреблением алкоголем?
<br>
<p style = "font-size: 16px"> Добавим индикатор злоупотребления алкоголем (Alcohol_indicator) 1 для злоупотребляющих и 0 для остальных. И уберем остальные столбцы.
</p>

```python
df_copy_alco = df_copy.copy()
df_copy_alco['Alcohol_indicator'] = ((df_copy_alco['Dalc'] > 3) | (df_copy_alco['Walc'] > 3)).astype(int)
drop_cols_alco = ['school', 'sex', 'age', 'address', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
df_copy_alco.drop(drop_cols_alco, axis=1, inplace=True)
freetime_analysis = df_copy_alco.groupby('freetime')['Alcohol_indicator'].mean()
print("Анализ по freetime:")
print(freetime_analysis)
goout_analysis = df_copy_alco.groupby('goout')['Alcohol_indicator'].mean()
print("\nАнализ по goout:")
print(goout_analysis)
```

<p style = "font-size: 16px"> Анализируя резульататы можно сказать что студенты у которых больше свободного время и частые встречи с друзьями больше склонны к злоупотреблению алкоголя.
</p>

### 14. Есть ли связь избытка свободного времени после школы и частых встреч с друзьями?
