<p style = "font-size: 16px">Использование matplotlib, подписей данных, заголовков. Использование простейших пользовательских цветов.</p>

```python
plt.scatter(data['age'], 
            data['balance'],
            color='red', edgecolors='navy')

plt.xlabel("Возраст клиента (age)")
plt.ylabel("Среднегодовой баланс (balance, евро)")
plt.title("Зависимость баланса на счёте от возраста клиента")
plt.show()
```

```python
# Раскрашивание данных
# Цвет в зависимости от диагноза (0 - нет подписки "no", 1 - есть подписка "yes")
c = data['y'].map({'no': 'red', 'yes': 'green'})
edge_c = data['y'].map({'no': 'darkred', 'yes': 'darkgreen'})

# Настройка графика
plt.scatter(data['age'],
            data['balance'],
            color=c, edgecolors=edge_c)

plt.xlabel("Возраст клиента (age)")
plt.ylabel("Среднегодовой баланс (balance, евро)")
plt.title("Распределение клиентов по возрасту и балансу с разделением по подписке на депозит")
plt.show()
```

```python
# Раскраска по диагнозу подписки
# 0 — нет подписки, 1 — есть подписка
data_no = data[data['y'] == 'no']
data_yes = data[data['y'] == 'yes']

# Получаем без подписки
plt.scatter(data_no['age'],
            data_no['balance'],
            color="green",
            edgecolors="lightgreen",
            label="Без подписки")

# Получаем с подпиской
plt.scatter(data_yes['age'],
            data_yes['balance'],
            color="red",
            edgecolors="darkred",
            label="С подпиской")

plt.xlabel("Возраст клиента (age)")
plt.ylabel("Среднегодовой баланс (balance, евро)")
plt.title("Распределение клиентов по возрасту и балансу с разделением по подписке")
plt.legend()
plt.show()
```

### 7. Корреляция признаков

<p style = "font-size: 16px">В реальных задачах машинного обучения при первичном анализе данных необходимо выявить корреляции признаков обучающей выборки. В пакете Pandas имеется встроенный инструмент для этого - метод corr() класса DataFrame. Ниже показан результат вывода этой функции.</p>

```python
data.corr(numeric_only=True)
```

<p style = "font-size: 16px">Полученная матрица имеет размер 7×7. Это незначительный размер (в реальных задачах матрица корреляций может иметь размер 100–1000 и более), но даже для матрицы рассматриваемого набора данных проанализировать корреляцию признаков вручную — трудоемкая задача. Можно использовать скрипты для выделения больших коэффициентов корреляции. Но лучше использовать специальный тип графика — heatmap (тепловую карту корреляций).</p>

```python
sns.heatmap(data.corr(numeric_only=True), cmap=plt.cm.PuBuGn)
plt.show()
```

<p style = "font-size: 16px">Коррелирующие признаки обычно удаляются или не рассматриваются в процессе обучения. Из карты heatmap видно, что некоторые признаки коррелируют: например, умеренная корреляция наблюдается в паре (pdays, previous) с коэффициентом около 0.45, а также слабые корреляции в парах (age, balance), (campaign, day). Отрицательные корреляции в основном слабые (например, pdays с campaign или day). Из таких пар можно удалить один признак</p>

```python
high_corr = ["previous", "pdays"]
data_uncorr = data.drop(high_corr, axis=1)
data_uncorr.columns
```

```python
sns.heatmap(data_uncorr.corr(numeric_only=True), cmap=plt.cm.rainbow)
plt.show()
```
